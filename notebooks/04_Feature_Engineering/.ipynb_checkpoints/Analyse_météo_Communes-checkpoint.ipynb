```python
# Chargement en mode colab
df = pd.read_csv(     "raw_2020_2024.csv" )
```

```python


# Mise en index temporel
df = df.rename(columns={"datetime_utc": "Date"})
df["Date"] = pd.to_datetime(df["Date"], utc=True)
df = df.set_index("Date").sort_index()

# =========================================================
# 2. Liste des communes représentatives
# =========================================================

communes = ["CRU", "SEL", "SVT", "BRA", "EYG"]

# =========================================================
# 3. Construction du tableau comparatif
# =========================================================

rows = []

for c in communes:
    row = {
        "commune": c,

        # --- Géographie / site ---
        "altitude_moy_m": df[f"{c}_altitude"].mean(),

        # --- Ressource solaire ---
        "ghi_moy": df[f"{c}_GHI"].mean(),
        "ghi_std": df[f"{c}_GHI"].std(),

        "clear_sky_ratio_moy": (
            df[f"{c}_GHI"] / df[f"{c}_Clear sky GHI"]
        ).mean(),

        # --- Vent ---
        "vent_moy": df[f"{c}_Vitesse_Vent"].mean(),
        "vent_std": df[f"{c}_Vitesse_Vent"].std(),

        # --- Nébulosité ---
        "nebulosite_moy": df[f"{c}_Nebulosite"].mean(),
        "nebulosite_std": df[f"{c}_Nebulosite"].std(),
    }

    rows.append(row)

df_communes_compare = pd.DataFrame(rows)

# =========================================================
# 4. Affichage et export
# =========================================================

display(df_communes_compare)

# Export
df_communes_compare.to_csv(
    "analyse_comparee_communes_paca.csv",
    index=False,
    encoding="utf-8"
)
```

```python
import matplotlib.pyplot as plt

communes = ["CRU", "SEL", "SVT", "BRA", "EYG"]

plt.figure(figsize=(12, 6))

for c in communes:
    monthly_ghi = (
        df
        .groupby(df.index.month)[f"{c}_GHI"]
        .sum()
    )

    plt.plot(
        monthly_ghi.index,
        monthly_ghi.values,
        marker="o",
        linewidth=2,
        label=c
    )

plt.title("Comparaison des communes représentatives – Irradiance mensuele cumulée")
plt.xlabel("Mois")
plt.ylabel("GHI mensuel cumulé (Wh/m²)")
plt.xticks(range(1, 13))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    df_communes_compare["altitude_moy_m"],
    df_communes_compare["ghi_moy"]
)

for _, row in df_communes_compare.iterrows():
    plt.annotate(
        row["commune"],
        (row["altitude_moy_m"], row["ghi_moy"])
    )

plt.xlabel("Altitude moyenne (m)")
plt.ylabel("GHI moyen")
plt.title("Lien altitude – ressource solaire")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()






```

```python
plt.figure(figsize=(12, 6))

for c in ["CRU", "SEL", "SVT", "BRA", "EYG"]:
    monthly_cloud = (
        df[f"{c}_Nebulosite"]
        .resample("ME")
        .mean()
        .groupby(lambda x: x.month)
        .mean()
    )

    plt.plot(
        monthly_cloud.index,
        monthly_cloud.values,
        marker="o",
        linewidth=2,
        label=c
    )

plt.title("Comparaison des communes – Nébulosité moyenne mensuelle")
plt.xlabel("Mois")
plt.ylabel("Nébulosité moyenne (%)")
plt.xticks(range(1,13))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

for c in ["CRU", "SEL", "SVT", "BRA", "EYG"]:
    monthly_wind = (
        df[f"{c}_Vitesse_Vent"]
        .resample("ME")
        .mean()
        .groupby(lambda x: x.month)
        .mean()
    )

    plt.plot(
        monthly_wind.index,
        monthly_wind.values,
        marker="o",
        linewidth=2,
        label=c
    )

plt.title("Comparaison des communes – Vitesse moyenne du vent (mensuelle)")
plt.xlabel("Mois")
plt.ylabel("Vitesse du vent moyenne (m/s)")
plt.xticks(range(1,13))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

communes = ["CRU", "SEL", "SVT", "BRA", "EYG"]

plt.figure(figsize=(12, 6))

for c in communes:
    monthly_temp = (
        df[f"{c}_Temperature"]
        .resample("ME")
        .mean()
        .groupby(lambda x: x.month)
        .mean()
    )

    plt.plot(
        monthly_temp.index,
        monthly_temp.values,
        marker="o",
        linewidth=2,
        label=c
    )

plt.title("Comparaison des communes – Température moyenne mensuelle")
plt.xlabel("Mois")
plt.ylabel("Température moyenne (°C)")
plt.xticks(range(1,13))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

communes = ["CRU", "SEL", "SVT", "BRA", "EYG"]

fig, axes = plt.subplots(
    nrows=3,
    ncols=1,
    figsize=(12, 14),
    sharex=True
)

# =========================================================
# 1️⃣ GHI mensuel cumulé
# =========================================================
ax = axes[0]

for c in communes:
    ghi_monthly = (
        df[f"{c}_GHI"]
        .resample("ME")
        .sum()
        .groupby(lambda x: x.month)
        .mean()
    )
    ax.plot(
        ghi_monthly.index,
        ghi_monthly.values,
        marker="o",
        linewidth=2,
        label=c
    )

ax.set_title("GHI mensuel cumulé")
ax.set_ylabel("GHI cumulé (Wh/m²)")
ax.grid(True, alpha=0.3)

# =========================================================
# 2️⃣ Nébulosité moyenne mensuelle
# =========================================================
ax = axes[1]

for c in communes:
    cloud_monthly = (
        df[f"{c}_Nebulosite"]
        .resample("ME")
        .mean()
        .groupby(lambda x: x.month)
        .mean()
    )
    ax.plot(
        cloud_monthly.index,
        cloud_monthly.values,
        marker="o",
        linewidth=2
    )

ax.set_title("Nébulosité moyenne mensuelle")
ax.set_ylabel("Nébulosité (%)")
ax.grid(True, alpha=0.3)

# =========================================================
# 3️⃣ Température moyenne mensuelle
# =========================================================
ax = axes[2]

for c in communes:
    temp_monthly = (
        df[f"{c}_Temperature"]
        .resample("ME")
        .mean()
        .groupby(lambda x: x.month)
        .mean()
    )
    ax.plot(
        temp_monthly.index,
        temp_monthly.values,
        marker="o",
        linewidth=2
    )

ax.set_title("Température moyenne mensuelle")
ax.set_ylabel("Température (°C)")
ax.set_xlabel("Mois")
ax.grid(True, alpha=0.3)

# =========================================================
# Légende commune
# =========================================================
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(
    handles,
    labels,
    loc="upper center",
    ncol=5,
    frameon=False
)

plt.xticks(range(1, 13))
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
