```python
def ajout_encodage_temporel_cyclique(df):
    """
    Ajoute des variables cycliques liées au temps :
    - heure de la journée : sin_hour, cos_hour
    - jour de l'année : sin_doy, cos_doy
    
    """

    # Copie pour éviter de modifier le DataFrame original
    df = df.copy()

    # Encodage cyclique de l'heure (heure + minutes)
    hour = df.index.hour + df.index.minute / 60
    df['sin_hour'] = np.sin(2 * np.pi * hour / 24)
    df['cos_hour'] = np.cos(2 * np.pi * hour / 24)

    # Encodage cyclique du jour de l'année
    # dayofyear commence à 1, donc on retire 1 pour démarrer à 0
    doy = df.index.dayofyear - 1

    # Nombre de jours selon que l'année est bissextile ou non
    nb_jour_annee = np.where(df.index.is_leap_year, 366, 365)

    df['sin_doy'] = np.sin(2 * np.pi * doy / nb_jour_annee)
    df['cos_doy'] = np.cos(2 * np.pi * doy / nb_jour_annee)

    return df

df_train = ajout_encodage_temporel_cyclique(df_train)
df_valid = ajout_encodage_temporel_cyclique(df_valid)
df_test  = ajout_encodage_temporel_cyclique(df_test)

display(df_train[['sin_hour', 'cos_hour', 'sin_doy', 'cos_doy']].head())

print("Taille de df_train :", df_train.shape)
print("Taille de df_valid :", df_valid.shape)
print("Taille de df_test  :", df_test.shape)
```

> L'`azimuth` représente la position horizontale du soleil (angle entre 0° et 360°).
C'est une variable **circulaire** :
> - 0° et 360° correspondent à la même direction
> - 359° est très proche de 1°
>
> Un encodage classique introduit donc une discontinuité artificielle.
Pour corriger cela, on transforme l'`azimuth` en **sinus** et **cosinus**, ce qui permet de préserver sa nature cyclique et d'améliorer les performances des modèles.

```python
def ajout_encodage_azimuth_cyclique(df, villes):
    """
    Ajoute un encodage cyclique de l'azimuth pour chaque ville :
    - {ville}_azimuth_sin
    - {ville}_azimuth_cos
    """

    # Copie pour éviter de modifier le DataFrame original
    df = df.copy()

    for ville in villes:
        col = f"{ville}_azimuth"

        # Vérifie que la colonne existe
        if col not in df.columns:
            continue

        # Conversion en radians
        azimuth_rad = np.radians(df[col])

        # Encodage cyclique
        df[f"{ville}_azimuth_sin"] = np.sin(azimuth_rad)
        df[f"{ville}_azimuth_cos"] = np.cos(azimuth_rad)

        # Suppression de la colonne brute
        df = df.drop(columns=[col])

    return df

df_train = ajout_encodage_azimuth_cyclique(df_train, villes)
df_valid = ajout_encodage_azimuth_cyclique(df_valid, villes)
df_test  = ajout_encodage_azimuth_cyclique(df_test, villes)

display(df_train[['bra_azimuth_sin', 'bra_azimuth_sin']].head())

print("Taille de df_train :", df_train.shape)
print("Taille de df_valid :", df_valid.shape)
print("Taille de df_test  :", df_test.shape)
```

```python
def appliquer_ponderation_regionale(df, col_par_ville, poids_par_ville):
    """
    Crée des colonnes régionales pondérées à partir des colonnes par ville,
    puis supprime les anciennes colonnes par ville.

    Paramètres
    ----------
    df : pd.DataFrame
        DataFrame d'entrée à transformer.
    col_par_ville : list
        Liste des variables à agréger, par exemple :
        ["ghi", "dni", "temp"]
    poids_par_ville : pd.Series
        Série contenant les poids, indexée par le préfixe de la ville.

    Retour
    ------
    pd.DataFrame
        Nouveau DataFrame transformé.
    """

    # Copie pour éviter de modifier le DataFrame original
    df = df.copy()

    # Dictionnaire qui contiendra les nouvelles colonnes régionales calculées
    region_cols = {}

    # Liste des colonnes par ville à supprimer après agrégation
    cols_to_drop = []

    # Liste des villes dans l'ordre des poids
    # villes = poids_par_ville.index.tolist()

    # Tableau numpy des poids pour le calcul vectorisé
    weights = poids_par_ville.to_numpy()

    # Boucle sur chaque variable à agréger
    for col in col_par_ville:
        # Construction des noms de colonnes attendues pour cette variable
        # Exemple : marseille_ghi, nice_ghi, toulon_ghi, ...
        all_concerns_cols = [f"{ville}_{col}" for ville in villes]

        # Vérification que toutes les colonnes nécessaires existent bien
        missing_cols = [c for c in all_concerns_cols if c not in df.columns]
        if missing_cols:
            raise KeyError(f"Colonnes manquantes dans df pour '{col}' : {missing_cols}")

        # Calcul de la somme pondérée :
        # chaque colonne ville_variable est multipliée par son poids,
        # puis on somme ligne par ligne
        region_cols[f"region_{col}"] = (
            df[all_concerns_cols]
            .mul(weights, axis=1)
            .sum(axis=1)
        )

        # On mémorise les colonnes d'origine à supprimer ensuite
        cols_to_drop.extend(all_concerns_cols)

    # Ajout en une seule fois des nouvelles colonnes régionales
    df = pd.concat([df, pd.DataFrame(region_cols, index=df.index)], axis=1)

    # Suppression des anciennes colonnes par ville
    df = df.drop(columns=cols_to_drop)

    # Copie finale optionnelle pour défragmenter le DataFrame
    df = df.copy()

    return df

df_region_train = appliquer_ponderation_regionale(df_train, col_par_ville, poids_par_ville)
df_region_valid = appliquer_ponderation_regionale(df_valid, col_par_ville, poids_par_ville)
df_region_test  = appliquer_ponderation_regionale(df_test,  col_par_ville, poids_par_ville)

print("Taille de df_region_train :", df_region_train.shape)
print("Taille de df_region_valid :", df_region_valid.shape)
print("Taille de df_region_test  :", df_region_test.shape)
```

```python
def construction_features_meteo(df, vars_meteo, lags, windows, prefix_list):
    new_cols = {}

    # Génération des variables dérivées pour chaque ville ou region et chaque variable météo
    for prefix in prefix_list:
        for var in vars_meteo:
            col_name = f"{prefix}{var}"
            if col_name not in df.columns:
                print(f"{col_name} n'existe pas.")

            serie = df[col_name]
            
            # Variation instantanée de la variable
            dserie = serie.diff()

            # Ajout de la différence première
            new_cols[f"{prefix}d{var}_dt"] = dserie

            # Création des variables retardées
            for lag in lags:
                new_cols[f"{prefix}{var}_lag_{lag}"] = serie.shift(lag)
                new_cols[f"{prefix}d{var}_lag_{lag}"] = dserie.shift(lag)

            # Calcul des statistiques glissantes
            for w in windows:
                new_cols[f"{prefix}{var}_roll_mean_{w}h"] = serie.rolling(w).mean()
                new_cols[f"{prefix}{var}_roll_std_{w}h"] = serie.rolling(w).std()

                new_cols[f"{prefix}d{var}_roll_mean_{w}h"] = dserie.rolling(w).mean()
                new_cols[f"{prefix}d{var}_roll_std_{w}h"] = dserie.rolling(w).std()

    return pd.concat([df, pd.DataFrame(new_cols, index=df.index)], axis=1)

prefix_villes = [f"{v}_" for v in villes]
prefix_region = ["region_"]

vars_meteo = ['ghi', 'nebulosite']
lags = [1, 2, 3]
windows = [4]

print("Shape avant construction_features_meteo:\n")
print(f"Taille de df_train : {df_train.shape}")
print(f"Taille de df_valid : {df_valid.shape}")
print(f"Taille de df_test  : {df_test.shape}\n")
print(f"Taille de df_region_train : {df_region_train.shape}")
print(f"Taille de df_region_valid : {df_region_valid.shape}")
print(f"Taille de df_region_test  : {df_region_test.shape}")

df_train = construction_features_meteo(df_train, vars_meteo, lags, windows, prefix_villes)
df_valid = construction_features_meteo(df_valid, vars_meteo, lags, windows, prefix_villes)
df_test  = construction_features_meteo(df_test, vars_meteo, lags, windows, prefix_villes)
df_region_train = construction_features_meteo(df_region_train, vars_meteo, lags, windows, prefix_region)
df_region_valid = construction_features_meteo(df_region_valid, vars_meteo, lags, windows, prefix_region)
df_region_test  = construction_features_meteo(df_region_test, vars_meteo, lags, windows, prefix_region)

print("Shape après construction_features_meteo:\n")
print(f"Taille de df_train : {df_train.shape}")
print(f"Taille de df_valid : {df_valid.shape}")
print(f"Taille de df_test  : {df_test.shape}\n")
print(f"Taille de df_region_train : {df_region_train.shape}")
print(f"Taille de df_region_valid : {df_region_valid.shape}")
print(f"Taille de df_region_test  : {df_region_test.shape}")
```

```python
def construction_features_ghi(df, prefixes):
    new_cols = {}

    # Calcul du maximum mensuel de la variable "solaire"
    # transform("max") permet de conserver un index aligné avec df
    solaire_monthly_max = df["solaire"].groupby(
        pd.Grouper(freq="ME")      #"ME" = new explicit name for month-end
    ).transform("max")

    for prefix in prefixes:
        # Extraction des variables GHI et de sa dérivée temporelle
        ghi = df[f"{prefix}ghi"]
        dghi = df[f"{prefix}dghi_dt"]
        clear = df[f"{prefix}clear_sky_ghi"]
        ghi_threshold = ghi.quantile(0.7)

        # Calcul du Clear Sky Index (CSI)
        # On évite la division par zéro avec clip
        csi = ghi / clear.clip(lower=1e-6)
        # On met à 0 les valeurs où le clear_sky est trop faible
        csi = csi.where(clear >= 1, 0)

        # Détection d'anomalies solaires (par mois)
        # Une anomalie est détectée si :
        # - irradiance élevée
        # - mais production solaire faible par rapport au max mensuel
        anomaly = (
            (ghi > ghi_threshold) & # Détection des périodes de forte irradiance
            (csi > 0.7) & # ciel clair
            (df["solaire"] < 0.9 * solaire_monthly_max) # production solaire faible par rapport au max mensuel
        ).astype(int)


        # Stockage des nouvelles variables dans un dictionnaire
        new_cols[f"{prefix}csi"] = csi
        new_cols[f"{prefix}abs_dghi_dt"] = dghi.abs()
        new_cols[f"{prefix}anomaly"] = anomaly

    return pd.concat([df, pd.DataFrame(new_cols, index=df.index)], axis=1)

df_train = construction_features_ghi(df_train, prefix_villes)
df_valid = construction_features_ghi(df_valid, prefix_villes)
df_test  = construction_features_ghi(df_test,  prefix_villes)

df_region_train = construction_features_ghi(df_region_train, prefix_region)
df_region_valid = construction_features_ghi(df_region_valid, prefix_region)
df_region_test  = construction_features_ghi(df_region_test,  prefix_region)

print("Shape après construction_features_ghi:\n")
print(f"Taille de df_train : {df_train.shape}")
print(f"Taille de df_valid : {df_valid.shape}")
print(f"Taille de df_test  : {df_test.shape}\n")
print(f"Taille de df_region_train : {df_region_train.shape}")
print(f"Taille de df_region_valid : {df_region_valid.shape}")
print(f"Taille de df_region_test  : {df_region_test.shape}")
```

```python
def construction_features_tch(df, lags, windows):
    new_cols = {}

    tch = df["tch_solaire"]
    dtch = tch.diff()
    abs_dtch = dtch.abs()    
    new_cols["dtch_dt"] = dtch
    new_cols["abs_dtch_dt"] = abs_dtch
    # Variables retardées du TCH solaire et de ses variations
    for lag in lags:
        new_cols[f"tch_lag_{lag}"] = tch.shift(lag)
        new_cols[f"dtch_lag_{lag}"] = dtch.shift(lag)
        new_cols[f"abs_dtch_lag_{lag}"] = abs_dtch.shift(lag)

    # Statistiques glissantes
    for w in windows:
        new_cols[f"tch_roll_mean_{w}h"] = tch.rolling(w).mean()
        new_cols[f"tch_roll_std_{w}h"] = tch.rolling(w).std()

        new_cols[f"dtch_roll_mean_{w}h"] = dtch.rolling(w).mean()
        new_cols[f"dtch_roll_std_{w}h"] = dtch.rolling(w).std()

        new_cols[f"abs_dtch_roll_mean_{w}h"] = abs_dtch.rolling(w).mean()
        new_cols[f"abs_dtch_roll_std_{w}h"] = abs_dtch.rolling(w).std()

    df_new = pd.DataFrame(new_cols, index=df.index)
    return pd.concat([df, df_new], axis=1)

df_train = construction_features_tch(df_train, lags, windows)
df_valid = construction_features_tch(df_valid, lags, windows)
df_test  = construction_features_tch(df_test, lags, windows)
df_region_train = construction_features_tch(df_region_train, lags, windows)
df_region_valid = construction_features_tch(df_region_valid, lags, windows)
df_region_test  = construction_features_tch(df_region_test,  lags, windows)

print("Taille après construction_features_tch:\n")
print(f"Taille de df_train : {df_train.shape}")
print(f"Taille de df_valid : {df_valid.shape}")
print(f"Taille de df_test  : {df_test.shape}\n")
print(f"Taille de df_region_train : {df_region_train.shape}")
print(f"Taille de df_region_valid : {df_region_valid.shape}")
print(f"Taille de df_region_test  : {df_region_test.shape}")
```
