```python
# Normalisation du nom des colonnes (minuscules, pas d'espace, pas de caractères spéciaux autre que '_')

# Récupération du nom des colunnes
nom_cols = df.columns

# Création d'une fonction pour normaliser le nom des colonnes
def clean_name(name):
    resultat = name.lower()
    resultat = resultat.replace('(', '')
    resultat = resultat.replace(')', '')
    resultat = resultat.replace('%', '')
    resultat = resultat.replace('.', '')
    resultat = resultat.strip()
    resultat = resultat.replace(' ', '_')
    resultat = resultat.replace('é', 'e')
    resultat = resultat.replace('è', 'e')
    return resultat

# Initialisation d'un dictionnaire
# les clés seront les noms initiaux
# les valeurs les noms normalisés
dico = {}

# Remplir le dictionnaire
for nom in nom_cols:
    dico[nom] = clean_name(nom)

# Normaliser le nom des colonnes
df = df.rename(columns=dico)
print(df.info())
df.tail()
```

```python
def add_cyclic_time_features(df):
    """
    Ajoute des variables cycliques liées au temps :
    - heure de la journée : sin_hour, cos_hour
    - jour de l'année : sin_doy, cos_doy

    Le DataFrame doit avoir un index de type datetime.
    """

    # Copie pour éviter de modifier le DataFrame original
    df = df.copy()

    # Encodage cyclique de l'heure (heure + minutes)
    hour = df.index.hour + df.index.minute / 60
    df['sin_hour'] = np.sin(2 * np.pi * hour / 24)
    df['cos_hour'] = np.cos(2 * np.pi * hour / 24)

    # Encodage cyclique du jour de l'année
    # dayofyear commence à 1, donc on retire 1 pour démarrer à 0
    doy = df.index.dayofyear - 1

    # Nombre de jours selon que l'année est bissextile ou non
    nb_jour_annee = np.where(df.index.is_leap_year, 366, 365)

    df['sin_doy'] = np.sin(2 * np.pi * doy / nb_jour_annee)
    df['cos_doy'] = np.cos(2 * np.pi * doy / nb_jour_annee)

    return df

df_train = add_cyclic_time_features(df_train)
df_valid = add_cyclic_time_features(df_valid)
df_test  = add_cyclic_time_features(df_test)

display(df_train[['sin_hour', 'cos_hour', 'sin_doy', 'cos_doy']].head())

print("Taille de df_train :", df_train.shape)
print("Taille de df_valid :", df_valid.shape)
print("Taille de df_test  :", df_test.shape)
```

```python
def create_meteo_features(df, vars_meteo, lags, windows, villes):
    new_cols = {}

    # Génération des variables dérivées pour chaque ville et chaque variable météo
    for ville in villes:
        for var in vars_meteo:
            serie = df[f"{ville}_{var}"]

            # Variation instantanée de la variable
            dserie = serie.diff()

            # Ajout de la différence première
            new_cols[f"{ville}_d{var}_dt"] = dserie

            # Création des variables retardées
            for lag in lags:
                new_cols[f"{ville}_{var}_lag_{lag}"] = serie.shift(lag)
                new_cols[f"{ville}_d{var}_lag_{lag}"] = dserie.shift(lag)

            # Calcul des statistiques glissantes
            for w in windows:
                new_cols[f"{ville}_{var}_roll_mean_{w}h"] = serie.rolling(w).mean()
                new_cols[f"{ville}_{var}_roll_std_{w}h"] = serie.rolling(w).std()

                new_cols[f"{ville}_d{var}_roll_mean_{w}h"] = dserie.rolling(w).mean()
                new_cols[f"{ville}_d{var}_roll_std_{w}h"] = dserie.rolling(w).std()

    df_new = pd.DataFrame(new_cols, index=df.index)
    return pd.concat([df, df_new], axis=1)

vars_meteo = ['ghi', 'bni', 'dhi', 'nebulosite']
lags = [1, 2, 3, 4, 6]
windows = [3, 4, 6]
villes = ['bra', 'cru', 'eyg', 'sel', 'svt']

print("Shape avant")
print(df_train.shape)
print(df_valid.shape)
print(df_test.shape)

df_train = create_meteo_features(df_train, vars_meteo, lags, windows, villes)
df_valid = create_meteo_features(df_valid, vars_meteo, lags, windows, villes)
df_test  = create_meteo_features(df_test, vars_meteo, lags, windows, villes)

print("Shape après")
print(f"Taille de df_train : {df_train.shape}")
print(f"Taille de df_valid : {df_valid.shape}")
print(f"Taille de df_test  : {df_test.shape}")
```

```python
def create_tch_features(df, lags, windows):
    new_cols = {}

    tch = df["tch_solaire"]
    dtch = tch.diff()
    abs_dtch = dtch.abs()

    # Variables retardées du TCH solaire et de ses variations
    for lag in lags:
        new_cols[f"tch_lag_{lag}"] = tch.shift(lag)
        new_cols[f"dtch_lag_{lag}"] = dtch.shift(lag)
        new_cols[f"abs_dtch_lag_{lag}"] = abs_dtch.shift(lag)

    # Statistiques glissantes
    for w in windows:
        new_cols[f"tch_roll_mean_{w}h"] = tch.rolling(w).mean()
        new_cols[f"tch_roll_std_{w}h"] = tch.rolling(w).std()

        new_cols[f"dtch_roll_mean_{w}h"] = dtch.rolling(w).mean()
        new_cols[f"dtch_roll_std_{w}h"] = dtch.rolling(w).std()

        new_cols[f"abs_dtch_roll_mean_{w}h"] = abs_dtch.rolling(w).mean()
        new_cols[f"abs_dtch_roll_std_{w}h"] = abs_dtch.rolling(w).std()

    df_new = pd.DataFrame(new_cols, index=df.index)
    return pd.concat([df, df_new], axis=1)

df_train = create_tch_features(df_train, lags, windows)
df_valid = create_tch_features(df_valid, lags, windows)
df_test  = create_tch_features(df_test, lags, windows)

print("Taille de df_train :", df_train.shape)
print("Taille de df_valid :", df_valid.shape)
print("Taille de df_test  :", df_test.shape)
```

```python
def apply_region_weighting(df, col_par_ville, poids_par_ville):
    """
    Crée des colonnes régionales pondérées à partir des colonnes par ville,
    puis supprime les anciennes colonnes par ville.

    Paramètres
    ----------
    df : pd.DataFrame
        DataFrame d'entrée à transformer.
    col_par_ville : list
        Liste des variables à agréger, par exemple :
        ["ghi", "dni", "temp"]
    poids_par_ville : pd.Series
        Série contenant les poids, indexée par le préfixe de la ville.

    Retour
    ------
    pd.DataFrame
        Nouveau DataFrame transformé.
    """

    # Copie pour éviter de modifier le DataFrame original
    df = df.copy()

    # Dictionnaire qui contiendra les nouvelles colonnes régionales calculées
    region_cols = {}

    # Liste des colonnes par ville à supprimer après agrégation
    cols_to_drop = []

    # Liste des villes dans l'ordre des poids
    villes = poids_par_ville.index.tolist()

    # Tableau numpy des poids pour le calcul vectorisé
    weights = poids_par_ville.to_numpy()

    # Boucle sur chaque variable à agréger
    for col in col_par_ville:
        # Construction des noms de colonnes attendues pour cette variable
        # Exemple : marseille_ghi, nice_ghi, toulon_ghi, ...
        all_concerns_cols = [f"{ville}_{col}" for ville in villes]

        # Vérification que toutes les colonnes nécessaires existent bien
        missing_cols = [c for c in all_concerns_cols if c not in df.columns]
        if missing_cols:
            raise KeyError(f"Colonnes manquantes dans df pour '{col}' : {missing_cols}")

        # Calcul de la somme pondérée :
        # chaque colonne ville_variable est multipliée par son poids,
        # puis on somme ligne par ligne
        region_cols[f"region_{col}"] = (
            df[all_concerns_cols]
            .mul(weights, axis=1)
            .sum(axis=1)
        )

        # On mémorise les colonnes d'origine à supprimer ensuite
        cols_to_drop.extend(all_concerns_cols)

    # Ajout en une seule fois des nouvelles colonnes régionales
    df = pd.concat([df, pd.DataFrame(region_cols, index=df.index)], axis=1)

    # Suppression des anciennes colonnes par ville
    df = df.drop(columns=cols_to_drop)

    # Copie finale optionnelle pour défragmenter le DataFrame
    df = df.copy()

    return df

df_train = apply_region_weighting(df_train, col_par_ville, poids_par_ville)
df_valid = apply_region_weighting(df_valid, col_par_ville, poids_par_ville)
df_test  = apply_region_weighting(df_test,  col_par_ville, poids_par_ville)

print("Taille de df_train :", df_train.shape)
print("Taille de df_valid :", df_valid.shape)
print("Taille de df_test  :", df_test.shape)
```
